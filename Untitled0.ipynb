```
import cv2
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import threshold_otsu
filter_image=io.imread(r"joint-replacement.jpg")
filter_image = cv2.cvtColor(filter_image, cv2.COLOR_RGB2GRAY)
entropy_image=entropy(filter_image, disk(5))
plt.imshow(entropy_image)
```

```
thresh=threshold_otsu(entropy_image)
binary=entropy_image <=thresh
plt.imshow(binary)
type(binary)
```
